```python
### Q1

def calculate_volume(r):
    
    # Ratio of a circle's perimeter over its diameter
    PI = 3.14159;
    
    # Return volume of sphere
    return 4/3 * PI * r ** 3


# Use function above for various test cases
print(calculate_volume(1))
print(calculate_volume(2))
```

```python
### Q2

def is_leap_year(y):
    
    # GO THRU FIRST EXCEPTION
    if y % 400 == 0:
        return True
    # then check for the False case
    elif y % 100 == 0:
        return False
    # then check the base True case
    elif y % 4 == 0:
        return True
    else:
        return False


# Use function above for various test cases
print(is_leap_year(2024))
print(is_leap_year(2022))
print(is_leap_year(2100))
print(is_leap_year(2000))
```

```python
### Q3

def get_longest_suffix(s1, s2):

    # This will contain the common characters, starting from the back of both strings
    suffix = ""

    # Start at the last character of both strings
    i1 = len(s1) - 1
    i2 = len(s2) - 1
    
    # Keep going until we go past the first character of either string
    while (i1 >= 0 and i2 >= 0):
        
        # If current characters are the same
        if not(s1[i1] == s2[i2]):
            
            # Don't add any more characters to the result
            return suffix
        
        # Otherwise, characters matched, so add onto the suffix
        suffix = s1[i1] + suffix
        
        # Go to the next character, further to the front of the strings
        i1 = i1 - 1
        i2 = i2 - 1
        
    # If we reach this point, then we ran out of characters in one of the strings,
    # and the entire shortest string can be found in the longest string
    return suffix


# Use above function for various test cases
print("'" + get_longest_suffix("station", "stallion") + "'")
print("'" + get_longest_suffix("firefighter", "fighter") + "'")
print("'" + get_longest_suffix("fighter", "firefighter") + "'")
print("'" + get_longest_suffix("python", "programming") + "'")
```

```python
### Q4

def swap_digit_pairs(n):
    
    # This will store the final result
    answer = 0
    
    # The current power of 10 for the digit that we are building,
    # e.g. if we are building last digit,        then 10^0
    #      if we are building second last digit, then 10^1
    #      if we are building third last digit,  then 10^2
    power_of_10 = 0
    
    # Keep going until no more digits
    while n != 0:
        
        # Get right digit in this pair of digits
        digit1 = n % 10
        
        # Remove last digit
        n = n // 10
        
        # If there are no more digits
        if n == 0:
            
            # Put digit in same position
            answer = answer + digit1 * 10**power_of_10
            
            # Get out of function immediately
            return answer
        
        # Get left digit in this pair of digits
        digit2 = n % 10
        
        # Remove last digit
        n = n // 10
        
        # Put digits in correct positions
        answer = answer + digit2 * 10**(power_of_10    )
        answer = answer + digit1 * 10**(power_of_10 + 1)
        
        # Get ready to see next pair of digits
        power_of_10 = power_of_10 + 2
        
    # Return final answer
    return answer


# Use function above for various test cases
print(swap_digit_pairs(123456))
print(swap_digit_pairs(12345))
```

```python
### Q5

def print_pattern(n):
    
    # Go through each row
    for r in range(1,2*n+1):
        
        # Go through each column
        for c in range(1,2*n+1):
            
            # 
            if r <= c and c <= n or r >= c and c > n or r+c >= 2*n+1 and r <= n or r+c <= 2*n+1 and r > n:
                print("*", end="")
                
            # Otherwise, print a space
            else:
                print(" ", end="")
                
        # Go to next line below
        print()
        
        
# Use function above
print_pattern(5)
print()
print_pattern(8)
```

```python
### Q6

def jumble_one_word(word):
    
    # If there no characters in the word, return empty string
    if len(word) == 0:
        return ""
    
    # This will contain the final jumbled word.
    # For now, we only get the first character
    result = word[0]
    
    # Go through each position in the word, except the first and last characters
    for i in range(1,len(word)-1):
        
        # Put current character in correct position in result word,
        # depending on whether we are in the front or back half of the result word.
        # If there are an odd number of characters, then the middle character is considered
        # to be in the back half
        if i < len(word)//2:
            result = result + word[2*i]
        else:
            result = result + word[2*(i - len(word)//2) + 1]
    
    # Get last charcater of word, if it has at least 2 characters,
    # because if it had only 1 charcter, then we already saw the last character
    # since it is also the first character
    if len(word) >= 2:
        result = result + word[len(word)-1]
    
    # Return final result
    return result


def jumble_words(s):
    
    # This list will contain words, all jumbled
    result = []
    
    # Position of start of word
    start = 0
    
    # Go through each position in the given string
    for i in range(len(s)):
        
        # If current character is a space
        if s[i] == " ":
            
            # Remember that this is the end of a word
            end = i
            
            # Get the previous word
            word = s[start:end]
            
            # Jumble this word
            jumbled_word = jumble_one_word(word)
            
            # Append onto end of list
            result.append(jumbled_word)
            
            # We are about to start a new word
            start = i + 1
            
    # Jumble the last word, which may not end in a space
    end = len(s)

    # Get the previous word
    word = s[start:end]

    # Jumble this word
    jumbled_word = jumble_one_word(word)

    # Append onto end of list
    result.append(jumbled_word)
    
    return result


# Test above function for various cases
print(jumble_words("Programming is really fun"))
print(jumble_words("Python is such an interesting language"))

result1 = jumble_words("Programming is really fun")
if result1[0] == 'Pormirgamng' and result1[1] == 'is' and result1[2] == 'ralely' and result1[3] == 'fun':
    print("test 1 passed")
else:
    print("test 1 FAILED ***")

result2 = jumble_words("Python is such an interesting language")
if result2[0] == 'Ptoyhn' and result2[1] == 'is' and result2[2] == 'scuh' and result2[3] == 'an' and result2[4] == 'itrsineetng' and result2[5] == 'lnugagae':
    print("test 2 passed")
else:
    print("test 2 FAILED ***")
```
