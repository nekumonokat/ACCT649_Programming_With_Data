```python
### Q2

def check_temps(a, b, c):

    # Did we go in any of the if statements?
    has_printed = False

    # Check if something is boiling
    if a >= 100 or b >= 100 or c >= 100:
        print("boiling")
        has_printed = True

    # Check if something is freezing
    if a <= 0 or b <= 0 or c <= 0:
        print("freezing")
        has_printed = True

    # Check if all equal
    if a == b and b == c:
        print("all equal")
        has_printed = True

    # Print message if we did not yet print
    if not(has_printed):
        print("normal")


# Use function above for various test cases
print("test 1")
print("boiling <- expected")
print("---")
check_temps(4, 102, 76)
print("==============================")

print("test 2")
print("freezing <- expected")
print("---")
check_temps(-76, 83, 4)
print("==============================")

print("test 3")
print("all equal <- expected")
print("---")
check_temps(76, 76, 76)
print("==============================")

print("test 4")
print("boiling <- expected")
print("freezing <- expected")
print("---")
check_temps(4, 102, -76)
print("==============================")

print("test 5")
print("boiling <- expected")
print("all equal <- expected")
print("---")
check_temps(176, 176, 176)
print("==============================")

print("test 6")
print("freezing <- expected")
print("all equal <- expected")
print("---")
check_temps(-76, -76, -76)
print("==============================")

print("test 7")
print("normal <- expected")
print("---")
check_temps(76, 83, 4)
print("==============================")
```

```python
### Q3

def match_positions(s1, s2):

    # Print s1
    print(s1)

    # Current positions in strings s1 and s2
    i1 = 0
    i2 = 0

    # Go through each position in each string until we reach end of either string
    while i1 < len(s1) and i2 < len(s2):

        # If current characters in the two strings match
        if s1[i1] == s2[i2]:

            # Print out character
            print(s1[i1], end="")

            # Search for next character in s2
            i2 += 1

        # Else, print space
        else:
            print(" ", end="")

        # Move on to next character in s1
        i1 += 1

    # Next time, print on next line below
    print()


# Use function above for various test cases
match_positions("Programming with Data", "gmwDa")
match_positions("Programming with Data", "gmwda")
match_positions("Programming with Data", "gmP")
match_positions("Programming with Data", "Xabc")
```

```python
### Q4
def swap_halves(n):

    # Remember the original n value
    original_n = n

    # How many digits in n?
    howManyDigits = 0

    # Keep going until no more digits
    while n > 0:

        # Count this digit
        howManyDigits += 1

        # Remove a digit
        n = n // 10

    # Reset original n value
    n = original_n

    # Right half of original n value
    right = 0

    # We will raise 10 to the power of something to put last digit in correct position
    powerOf10 = 0

    # Keep going until we remove half the digits
    for i in range(howManyDigits // 2):

        # Get last digit
        lastDigit = n % 10

        # Put last digit in correct position
        right += lastDigit * 10**powerOf10

        # Get ready to put next digit in correct position
        powerOf10 += 1

        # Remove a digit
        n = n // 10

    # Combine left and right parts by swapping them.
    # If there are an even number of digits
    if howManyDigits % 2 == 0:
        answer = right * 10**powerOf10 + n

    # Else, we have odd number of digits
    else:

        # Get last digit of n, which will be the new middle digit
        middle = n % 10

        # Remove last digit from n
        n = n // 10                

        # Combine the parts in correct positions
        answer = right * 10**(powerOf10+1) + middle * 10**powerOf10 + n
    return answer
    

# Use above function
print(swap_halves(123456))
print(swap_halves(12345))
```

```python
### Q5

def print_pattern(n):
    
    # Go through each row
    for r in range(1,2*n+1):
        
        # Go through each column
        for c in range(1,2*n+1):
            
            # 
            if r + c <= n + 1 or c >= n + r or (r > n and c >= r - (n-1) and r + c < 3*n + 1):
                print("*", end="")
                
            # Otherwise, print a space
            else:
                print(" ", end="")
                
        # Go to next line below
        print()
        
        
# Use function above
print_pattern(5)
print()
print_pattern(8)
```

```python
### Q6
def increase_integers(s, n):

    # This list will store all the integers to be returned
    list = []

    # All possible digits
    digits = "1234567890"

    # Value of current integer
    currentInteger = 0

    # Was the previous character a digit?
    previousWasDigit = False

    # Go through each character in string
    for c in s:

        # If current character is a digit
        if c in digits:

            # Add this digit onto end of current character
            currentInteger = 10 * currentInteger + int(c)

            # Current character is digit, so it is about to become previous character
            previousWasDigit = True

        # Else, current character is not a digit
        else:

            # If previous character was digit
            if previousWasDigit:

                # Add n to current integer
                currentInteger += n

                # Put the integer into the list
                list.append(currentInteger)

                # Reset integer, since we ended the integer
                currentInteger = 0

            # Current character is not digit, so it is about to become previous character
            previousWasDigit = False

    # If last character was digit, then put current character in list
    if previousWasDigit:
        currentInteger += n
        list.append(currentInteger)
    
    return list
            


# Use above function
print(increase_integers("abc 123 defgh 456 ij", 91))
print(increase_integers("abc 123 defgh 00456 ij", 91))
print(increase_integers("abc 123de fgh456 ij", 91))
print(increase_integers("abc 123 defgh 456 ij 78", 91))
```
