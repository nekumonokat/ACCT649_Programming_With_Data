```python
### Q2

def check_earlier_in_year(d1, m1, d2, m2):

    # Compare months
    if m1 < m2:
        return 1
    if m1 > m2:
        return 2

    # Compare days
    if d1 < d2:
        return 1
    if d1 > d2:
        return 2

    # If we reach here, that means both people have same birthday (ignoring year)
    return 0


# Convert from which day is earlier in the year to who is younger
def earlier_to_younger(n):
    if n == 1:
        return 2
    elif n == 2:
        return 1
    else:
        return 0


def check_younger(d1, m1, y1, d2, m2, y2):

    # Compare years
    if y1 > y2:
        return 1
    if y1 < y2:
        return 2

    # If year is same, then compare months and days
    earlier = check_earlier_in_year(d1, m1, d2, m2)

    # Whichever birthday is earlier means they are older, so we have to switch it
    return earlier_to_younger(earlier)


def check_birthdays(d1, m1, y1, d2, m2, y2):

    # Who is younger?
    younger = check_younger(d1, m1, y1, d2, m2, y2)

    # Who is born earlier in year?
    earlier = check_earlier_in_year(d1, m1, d2, m2)

    # If exactly same age and birthday, including year
    if younger == 0:
        print("Both have exactly same age")

    # If same birthday but not on same year
    elif earlier == 0:
        print("Person " + str(younger) + " is younger but they both have same birthday")

    # Else, see if younger and earlier is same or not
    else:
        if younger == earlier:
            print("Person " + str(younger) + " is younger and has their birthday earlier in the year")
        else:
            print("Person " + str(younger) + " is younger but Person " + str(earlier) + " has their birthday earlier in the year")


# Use function above for various test cases
# check_birthdays(23, 9, 2025, 5, 9, 2025)
# check_birthdays(5, 9, 2025, 23, 9, 2025)
# check_birthdays(23, 9, 2025, 5, 2, 2025)
# check_birthdays(23, 9, 1999, 5, 2, 2025)
# check_birthdays(23, 9, 1999, 23, 9, 2025)
# check_birthdays(23, 9, 2025, 23, 9, 2025)

check_birthdays(10, 4, 2025, 2, 4, 2025)   # 1a - same month
check_birthdays(16, 8, 2025, 11, 3, 2025)  # 2a - same year
check_birthdays(5, 7, 2002, 3, 2, 1990)    # 3a - not same year
check_birthdays(2, 4, 2025, 10, 4, 2025)   # 1b
check_birthdays(11, 3, 2025, 16, 8, 2025)  # 2b
check_birthdays(3, 2, 2002, 5, 7, 1990)    # 3b
check_birthdays(31, 1, 2025, 31, 1, 1999)  # 4a - same birthday, diff year
check_birthdays(31, 1, 1999, 31, 1, 2025)  # 4b
check_birthdays(1, 3, 2000, 1, 3, 2000)    # 5  - same birthday, incl year
```

```python
### Q3

def add_digits(s):

    # Sum of all digits in string
    total = 0

    # All possible digits
    digits = "1234567890"

    # Go through each character in the given string
    for c in s:

        # If current character is a digit
        if c in digits:

            # First, convert character into integer
            int_value = int(c)

            # Add integer value to total
            total += int_value

    # Give back final answer to whoever used the function
    return total


# Use function above for various test cases
# print(add_digits("abc621de4f"))
# print(add_digits("abc-621de.4f"))
# print(add_digits("abcde"))

print(add_digits("mfs7e2oj8aski4ya"))
print(add_digits("ojvs26oa7webw73rha94aiuav"))
print(add_digits("abcde"))
```

```python
### Q4

def get_smallest_pair(n):

    # If less than 2 digits, then there is no smallest pair
    if n < 10:
        return -1

    # Smallest pair of digits found so far
    min_pair = 100

    # Keep going until no more pair of digits
    while n >= 10:

        # Get last 2 digits
        pair = n % 100

        # If this pair of digits is smallest so far, remember it
        if pair < min_pair:
            min_pair = pair

        # Remove last digit
        n //= 10

    # Return smallest pair found to whoever use this function
    return min_pair


def remove_smallest_pair(n):

    # Determine value of smallest pair
    min_pair = get_smallest_pair(n)

    # If there is no pair of digits, then don't change given number
    if min_pair == -1:
        return n

    # This will store the final answer, with the smallest pair of digits removed
    answer = 0

    # The position of the next digit that we will write into answer, e.g.
    # 0 => 10^0 (right-most digit)
    # 1 => 10^1 (second right-most digit)
    # 2 => 10^2 etc...
    exponent = 0
    
    # Keep going until no more digits
    while n > 0:

        # Get last 2 digits
        pair = n % 100

        # If this pair of digits is the right-most occurrence of the smallest pair
        if pair == min_pair:

            # Remove this pair of digits from n
            n //= 100

            # Put all remaining digits of n into correct position in answer
            answer += n * 10**exponent

            # Get out of function immediately
            return answer

        # Else, get last digit
        else:
            digit = n % 10
    
            # Put digit in correct position in answer
            answer += digit * 10**exponent
    
            # Get ready to write into next position next time
            exponent += 1
            
        # Remove last digit
        n //= 10


# Use function above for various test cases
# print(remove_smallest_pair(17812345))
# print(remove_smallest_pair(71234125))
# print(remove_smallest_pair(340915))
# print(remove_smallest_pair(9))
# print(remove_smallest_pair(10))
# print(remove_smallest_pair(-17812345))

print("########### Test 1 ###########")
print("Expected: 986496684")
print("Actual:   " + str(remove_smallest_pair(98649629684)))

print("########### Test 2 ###########")
print("Expected: 982964668")
print("Actual:   " + str(remove_smallest_pair(98296429668)))

print("########### Test 3 ###########")
print("Expected: 929649668")
print("Actual:   " + str(remove_smallest_pair(90829649668)))

print("########### Test 4 ###########")
print("Expected: 7")
print("Actual:   " + str(remove_smallest_pair(7)))

print("########### Test 5 ###########")
print("Expected: 0")
print("Actual:   " + str(remove_smallest_pair(37)))
```

```python
### Q5

def print_pattern(n):
    
    # Go through each row
    for r in range(1,2*n+1):
        
        # Go through each column
        for c in range(1,3*n+1):
            
            # If we are in first triangle, or second triangle, or third triangle, or at bottom rectangle,
            # and we are not in the hole at the bottom, print a star
            if ((r+c >= n+1 and r <= n and c <= n) or (r+c >= 2*n+1 and r <= n and c <= 2*n) or (r+c >= 3*n+1 and r <= n and c <= 3*n) or (r > n)) and not (r > 3*n/2 and c > n and c <= 2*n):
                print("*", end="")
                
            # Otherwise, print a space
            else:
                print(" ", end="")
                
        # Go to next line below
        print()
        
        
# Use function above for various test cases
# print_pattern(4)
# print()
# print_pattern(6)

print_pattern(8)
print()
print_pattern(2)
```

```python
### Q6

def get_average(orig_list, i):

    # If first element, there is no previous element
    if i == 0:

        # If there is a next element, use it to calculate average
        if len(orig_list) > 1:
            return (orig_list[i] + orig_list[i+1]) / 2

        # Else, there's no next element, so just return only element
        else:
            return orig_list[i]

    # If last element, there is no next element
    elif i == len(orig_list)-1:
        return (orig_list[i-1] + orig_list[i]) / 2

    # Else, there is a previous and next element
    else:
        return (orig_list[i-1] + orig_list[i] + orig_list[i+1]) / 3


def average_list(orig_list):

    # This will contain the final result, with each value being the average of its surrounding values
    result_list = []

    # Go through each position in the original list
    for i in range(len(orig_list)):

        # Calculate average of numbers in list, centered at current position i
        average = get_average(orig_list, i)

        # Put onto end of result list
        result_list.append(average)

    # Copy values from result list to orig list
    for i in range(len(result_list)):
        orig_list[i] = result_list[i]


# Use function above for various test cases
numbers = [3,7,4,5,6,9,3,4,7,8]
average_list(numbers)
print(numbers)

numbers = [3,7,4,5,6]
average_list(numbers)
print(numbers)
```

```python
### Q6

# Returns the average of the "n" elements centered at index "center" in the list "orig_list"
def get_average(orig_list, n, center):

    # Sum of integers that we are averaging
    total = 0

    # How many number we are averaging
    count = 0

    # Go through each of the n possible positions
    # of numbers that we may have to average
    for i in range(center - n//2, center + n//2 + 1):

        # If this position is valid
        if i >= 0 and i < len(orig_list):

            # Add this number to total
            total += orig_list[i]

            # Count this value
            count += 1

    # Return average
    return total / count


def average_list(orig_list, n):

    # This will contain the final result, with each value being the average of its surrounding values
    result_list = []

    # Go through each position in the original list
    for i in range(len(orig_list)):

        # Calculate average of n numbers in list, centered at current position i
        average = get_average(orig_list, n, i)

        # Put onto end of result list
        result_list.append(average)

    # Copy values from result list to orig list
    for i in range(len(result_list)):
        orig_list[i] = result_list[i]


# Use function above for various test cases
# numbers = [9,7,4,5,6]
# average_list(numbers, 3)
# print(numbers)

# numbers = [9,7,4,5,6,7]
# average_list(numbers, 5)
# print(numbers)

numbers = [4,5,6,10,1,4,0]
average_list(numbers, 3)
print("Expected: [4.5, 5.0, 7.0, 5.666666666666667, 5.0, 1.6666666666666667, 2.0]")
print("Actual:   ", end="")
print(numbers)
print()

numbers = [4,5,6,10,1,4,0,2,9,4]
average_list(numbers, 7)
print("Expected: [6.25, 5.2, 5.0, 4.285714285714286, 4.0, 4.571428571428571, 4.285714285714286, 3.3333333333333335, 3.8, 3.75]")
print("Actual:   ", end="")
print(numbers)
```
