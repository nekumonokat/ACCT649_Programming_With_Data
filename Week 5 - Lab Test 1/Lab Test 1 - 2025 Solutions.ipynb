```python
### Q2: 20/20

def get_later_in_year(d1, m1, d2, m2):
    # birth month is later
    if m1 > m2: return 1
    elif m2 > m1: return 2
    
    # if month same, check day
    else:
        # birth day is later
        if d1 > d2: return 1
        elif d2 > d1: return 2
        # same birthdays, different years
        else: return 0

def get_younger(d1, m1, y1, d2, m2, y2):
    # birth year is later
    if y1 > y2: return 1
    elif y2 > y1: return 2
    
    # if year same, check month
    else: return get_later_in_year(d1, m1, d2, m2)

def get_earlier_in_year(d1, m1, d2, m2):
    if get_later_in_year(d1, m1, d2, m2) == 1: return 2
    elif get_later_in_year(d1, m1, d2, m2) == 2: return 1
    else: return 0

def check_birthdays(d1, m1, y1, d2, m2, y2):

    # 1. FINDING YOUNGER PERSON
    younger = get_younger(d1, m1, y1, d2, m2, y2)
    # 2. FINDING EARLIER BIRTHDAY
    earlier = get_earlier_in_year(d1, m1, d2, m2)

    # CONDITION 1: SAME BIRTHDAY SAME AGE:
    if younger == 0:
        print("Both have exactly same age")
    # CONDITION 2: DIFFERENT AGE BUT SAME BIRTHDAY:
    elif earlier == 0:
        print(f"Person {younger} is younger but they both have same birthday")
    # CONDITION 3: IF YOUNGER AND EARLIER:
    elif younger == earlier:
        print(f"Person {younger} is younger and has their birthday earlier in the year")
    # CONDITION 4: ONE YOUNGER BUT OTHER EARLIER:
    elif younger != earlier:
        print(f"Person {younger} is younger but Person {earlier} has their birthday earlier in the year")

# Use function above for various test cases
check_birthdays(23, 9, 2025, 5, 9, 2025)
check_birthdays(5, 9, 2025, 23, 9, 2025)
check_birthdays(23, 9, 2025, 5, 2, 2025)
check_birthdays(23, 9, 1999, 5, 2, 2025)
check_birthdays(23, 9, 1999, 23, 9, 2025)
check_birthdays(23, 9, 2025, 23, 9, 2025)
```

```python
### Q3: 20/20

def add_digits(string):
    
    # sum of digits seen
    total = 0
    digits = "1234567890"

    # go thru each character in string
    for char in string:
        # if current char is a digit
        if char in digits:
            total += int(char)

    return total

# Use function above for various test cases
print(add_digits("abc621de4f"))
print(add_digits("abc-621de.4f"))
print(add_digits("abcde"))
```

```python
### Q4: 14/20

def get_min_pair(n):
    # a pair cannot exceed 99
    min_pair = 100 
    
    # REPEATEDLY % AND //
    # keep going until no more pairs of digits
    while n >= 10:
        pair = n % 100

        # checks if current pair is smallest
        if pair < min_pair: min_pair = pair
        
        # remove last digit
        n //= 10

    return min_pair
    

def remove_smallest_pair(n):

    # store smallest pair
    min_pair = get_min_pair(n)
    # this will accumulate
    answer = 0
    # stores the power
    power = 0
    # stores if removal has been done
    is_removed = False

    # keep going until no more pairs of digits
    while n >= 10:
    
        # getting last 2 digits
        pair = n % 100

        # check if last 2 digits are minimum pair
        if pair == min_pair and not is_removed:
            # removes 1 digit in n (end of loop removes another digit)
            n //= 10
            is_removed = True

        else:
            last_digit = n % 10
            # adding to answer with correct power
            answer += last_digit * (10**power)
            # incrementing power
            power += 1

        # remove last digit
        n //= 10

    answer += n * (10**power)
    return answer

# Use function above for various test cases
print(remove_smallest_pair(17812345))
print(remove_smallest_pair(71234125))
print(remove_smallest_pair(340915))
print(remove_smallest_pair(9))
print(remove_smallest_pair(10))
```

```python
### Q5: 20/20

def print_pattern(n):
    for r in range(1, 2*n+1):
        for c in range(1, 3*n+1):

            # TOP HALF OF ROWS
            # first triangle
            if r+c >= n+1 and r <= n and c <= n:
                print("*", end = "")
            # second triangle
            elif r+c >= 2*n+1 and r <= n and c <= 2*n:
                print("*", end = "")
            # third triangle
            elif r+c >= 3*n+1 and r <= n and c <= 3*n:
                print("*", end = "")

            # BOTTOM HALF OF ROWS
            # this logic just prints everywhere but the hole
            elif r > n and not (r > 1.5*n and c > n and c <= 2*n):
                print("*", end = "")

            else:
                print(" ", end = "")

        print()

# Use function above for various test cases
print_pattern(4)
print()
print_pattern(6)
```

```python
### Q6: 16/20

def get_avg(old_list, n, center):
    # sum of values to be averaged
    total = 0
    # depending on where you are, count
    count = 0

    # go thru each possible position to count in the average
    for i in range(center - n//2, center + n//2 + 1):
        # checking index is in range
        if i >= 0 and i < len(old_list):
            total += old_list[i]
            count += 1

    return total / count

get_avg([9,7,4,5,6], 3, 4)

def average_list(numbers, n):
    # this contains the final answer
    new_list = []

    # go thru each position in the list
    for i in range(len(numbers)):
        # append to new_list
        new_list.append(get_avg(numbers, n, i))

    # change the respective numbers element to new_list element
    for i in range(len(numbers)):
        numbers[i] = new_list[i]

# Use function above for various test cases
numbers = [9,7,4,5,6]
average_list(numbers, 3)
print(numbers)

numbers = [9,7,4,5,6,7]
average_list(numbers, 5)
print(numbers)
```
