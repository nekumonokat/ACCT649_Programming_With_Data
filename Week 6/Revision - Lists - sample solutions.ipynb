```python
### Q1

def reverse_words(s):
    
    # This will contain the final result
    result = ""
    
    # Get a list of all words, which are separated by spaces
    words = s.split()
    
    # Is this the first word
    first_word = True
    
    # Go through each word
    for w in words:
        
        # Reverse the word
        reversed_word = w[len(w)-1::-1]
        
        # Append onto result
        if (first_word):
            result = reversed_word
            first_word = False
        else:
            result = result + " " + reversed_word
    
    # Return final result
    return result


# Use function above
print("'" + reverse_words("I study at SMU") + "'")
print("'" + reverse_words("Python is a programming language.") + "'")
print("'" + reverse_words("") + "'")
```

```python
### Q2
def find_greatest_common_divisor(a, b):
    
    # Assume a is less than b. If not, swap them.
    if (b < a):
        temp = a
        a = b
        b = temp
        
    # Go through each integer, from 'a' down to 1 exclusive
    for i in range(a, 1, -1):
        
        # If current integer is a divisor of both 'a' and 'b'
        if a % i == 0 and b % i == 0:
            return i
        
    # If we reach here, that means no integer above 1 is a common divisor,
    # and 1 is a common divsor of all integers
    return 1

# Use above function
print(find_greatest_common_divisor(8, 12))
print(find_greatest_common_divisor(100, 250))
print(find_greatest_common_divisor(13, 25))
```

```python
### Q3

def get_average_duration(movies):

    # Total time of movies
    total = 0
    
    # If list is empty
    if len(movies) == 0:
        return -1
    
    # Go through each movie
    for m in movies:
        
        # Add movie duration to total
        total += m[2]
        
    # Return average
    return total / len(movies)


def get_num_movies_of_genre(movies, genre):

    # How many movies of this genre?
    count = 0
    
    # Go through each movie
    for m in movies:
        
        # If this movie is in the target genre
        if m[1] == genre:
            
            # Count it
            count = count + 1
            
    # Return final count
    return count


def get_title_of_longest_movie(movies):
    
    # If list is empty
    if len(movies) == 0:
        return ""

    # Assume first movie is longest movie so far
    longest_movie = movies[0]
    
    # Go through each movie index
    for i in range(1, len(movies)):
        
        # If current movie has longer duration than longest movie so far
        if movies[i][2] > longest_movie[2]:
        
            # Remember this movie
            longest_movie = movies[i]
        
    # Return longest movie
    return longest_movie


def get_movies_with_keyword(movies, keyword):
    
    # This list will store movies with the given keyword in title
    result = []
    
    # Go through each movie
    for m in movies:
        
        # If keyword is in movie title
        if m[0].find(keyword) != -1:
            
            # Append this movie to result list
            result.append(m)
            
    # Return final count
    return result
    
    
# Create a list of movies.
# Each movie is a tuple containing the title, genre, and duration of movie in minutes.
movies = [("The Shawshank Redemption", "drama", 142), 
        ("The Godfather", "drama", 175), 
        ("The Lord of the Rings: The Return of the King", "fantasy", 201), 
        ("Forrest Gump", "comedy", 142), 
        ("The Lord of the Rings: The Fellowship of the Ring", "fantasy", 178), 
        ("Star Wars: Episode IV - A New Hope", "fantasy", 121), 
        ("The Lion King", "animation", 88)]

# What is the movie average duration? 
print(get_average_duration(movies))

# How many drama movies?
print(get_num_movies_of_genre(movies, "drama"))

# How many fantasy movies?
print(get_num_movies_of_genre(movies, "fantasy"))

# Which movie has longest duration?
print(get_title_of_longest_movie(movies))

# Which movies have "King" in the title?
print(get_movies_with_keyword(movies, "King"))

# Which movies have "Lord" in the title?
print(get_movies_with_keyword(movies, "Lord"))
```

```python
### Q4

def is_part_of_number(c):
    return c in "1234567890."

def compute_avg_height(s):
    
    # Total of heights
    total = 0
    
    # How many heights
    count = 0;
    
    # Are we currently in a number?
    currently_in_number = False
    
    # Current number, expressed as a string
    current_number_string = ""
    
    # Go through each character in given string
    for c in s:
        
        # If current character is part of a number
        if is_part_of_number(c):
            
            # Remember that we are currently in number
            currently_in_number = True
            
            # Append current character onto current number expressed as a string
            current_number_string = current_number_string + c
            
        # Else, if we have just ended as string
        elif currently_in_number:
            
            # Remember that we are no longer currently in number
            currently_in_number = False
            
            # Add current number to total
            total = total + float(current_number_string)
            
            # Count this number
            count = count + 1
            
            # Reset current number expressed as a string
            current_number_string = ""

    # Calculate average
    return total / count
            
    
# Use above function
compute_avg_height("Jonathan Li:1.75m, Lim, Josephine : 1.59m , George Khoo:   1.7 m")
```

```python
### Q5a

def get_larger_values(numbers):
    
    # Sum of elements in list
    total = 0
    
    # Go through each number
    for n in numbers:
        
        # Add current number to total
        total = total + n
        
    # Calculate average
    average = total / len(numbers)
    
    # This list will contain numbers greater than average
    result = []
    
    # Go through each number
    for n in numbers:
    
        # If current number is greater than average
        if n > average:
            
            # Add current element into list
            result.append(n)
    
    # Return final list
    return result

    
# Use above function
get_larger_values([2.5, 3.5, 5.5, 1.0])
```

```python
### Q5b

def merge_list(numbers1, numbers2):
    
    # This list will contain elements from both lists
    result = []
    
    # Current position in both lists
    i = 0
    
    # Keep going while we have not yet gone past end of both lists
    while (i < len(numbers1) or i < len(numbers2)):
        
        # If we have not yet past end of first list
        if i < len(numbers1):
            
            # Append current element onto result list
            result.append(numbers1[i])
        
        # If we have not yet past end of first list
        if i < len(numbers2):
            
            # Append current element onto result list
            result.append(numbers2[i])
        
        # Go to next position
        i = i + 1
    
    # Return merged list
    return result

    
# Use above function
merge_list([1, 3, 10, 15, 4, 7, 12], [9, 5, 2]) 
```

```python
### Q5c

def check_numbers(numbers1, numbers2):
    
    # Go through each integer in first list
    for n1 in numbers1:
        
        # Is n1 divisible by a number in second list?
        divisible = False
        
        # Go through each integer in second list
        for n2 in numbers2:
            
            # If number in first list is divisible by number in second list
            if n1 % n2 == 0:
                divisible = True
                
        # If n1 is not divisible by a number in second list
        if not divisible:
            return False
        
    # If we reach here, that means all numbers in first list are divisible by at least one number in second list
    return True


# Use above function
print(check_numbers([6, 15, 9], [2, 3]))
print(check_numbers([6, 15, 9], [2, 4]))
```

```python
### Q6a

def get_unique_titles(books):
    
    # This list will contain boolean values,
    # corresponding to whether the book in the same position is unique
    is_first_list = []
    for i in range(len(books)):
        is_first_list.append(True)

    # Index of book we are looking at
    i = 0
    
    # Go through each book
    while i < len(books):
        
        # Index of book we are comparing to
        j = i + 1
        
        # If this book is the first with this title
        if is_first_list[i]:
        
            # Go through each other book that we will compare with
            while j < len(books):

                # If the book is not being compared to itself,
                # and the book titles are the same
                if i != j and books[i][0] == books[j][0]:

                    # Remember that the second of this book title is not the first
                    is_first_list[j] = False

                # Go to next book
                j = j + 1
        
        # Go to next book
        i = i + 1
        
    # This list will contain unique titles
    result = []
    
    # Go through each boolean value in is_unique_list
    for i in range(len(is_first_list)):
        
        # If this book is the first book with that title
        if is_first_list[i]:
            
            # Append this book title to result list
            result.append(books[i][0])
    
    # Return list of unique book titles
    return result

    
# Use above function
get_unique_titles([("Intro to Programming", "Ed-2", "paperback", 2), ("Intro to Python", "Ed-1", "paperback", 5), ("Intro to Programming", "Ed-3", "hardcover", 4)]) 
```

```python
### 6b

def get_titles_and_counts(books):

    # Create a dictionary to store book title - count (string-integer) pairs
    dictionary = {}
    
    # Go through each book
    for b in books:
        
        # Get book title
        book_title = b[0]
        
        # If this book title is in dictionary
        if book_title in dictionary:
            
            # Get the old count from dictionary
            old_count = dictionary[book_title]
            
            # Put the new count into dictionary
            dictionary[book_title] = old_count + b[3]
            
        # Else, put this book title and its count into dictionary
        else:
            dictionary[book_title] = b[3]
        
    # This list will store tuples containing book titles and counts
    result = []
            
    # Go through each book title
    for b in dictionary.keys():
        
        # Create a tuple containing book title and count
        book_tuple = (b, dictionary[b])
        
        # Append tuple to list
        result.append(book_tuple)
        
    # Return final result
    return result
        
    
# Use above function
get_titles_and_counts([("Intro to Programming", "Ed-2", "paperback", 2), ("Intro to Python", "Ed-1", "paperback", 5), ("Intro to Programming", "Ed-3", "hardcover", 4), ("Intro to Python", "Ed-3", "hardcover", 3)])
```

```python
### Q7

def retrieve_numbers(s):
    
    # This string will containg numbers, separated by spaces
    result = ""
    
    # The current number, represented as a string
    current_number_string = ""
    
    # Go through each character in given string
    for c in s:
        
        # If current character is a digit
        if c in "1234567890":
            
            # Append digit to current number
            current_number_string = current_number_string + c
            
        # Else, if current number is not blank
        elif current_number_string != "":
            
            # Append current number to result
            if result == "":
                result = current_number_string
            else:
                result = result + " " + current_number_string
            
            # Reset current number
            current_number_string = ""
            
    # If current number is not blank
    if current_number_string != "":
            
        # Append current number to result
        if result == "":
            result = current_number_string
        else:
            result = result + " " + current_number_string
            
    # Return final result
    return result


# Use above function
print("'" + retrieve_numbers("12abc600$##0900AB 100X")  + "'")
print("'" + retrieve_numbers("34.5689abc980")   + "'")
print("'" + retrieve_numbers("xyz") + "'")
print("'" + retrieve_numbers("abc25xyz")   + "'")
```
