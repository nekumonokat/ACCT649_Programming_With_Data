## GroupBy

The groupby method can be used to group a set of rows together based on a criteria. We can use columns as parameters to separate/group rows. For example, in the `tips` dataset, you can perform groupby operations if you want to see the income dynamics across various days ('day' column) or time of the day ('time' column).
See the example below:

```python
# Group by time
tips_time = df_tips.groupby('time')

# Group by day
tips_day = df_tips.groupby('day')

# Write your code here to examine the data type of tips_time and tips_day
print(type(tips_time))
print(type(tips_day))

# Use the head(4) on the groupby objects to see how its different to using head(4) on the original dataset
print(tips_time.head(4))
print("")
print(tips_day.head(4))
print("")
print(df_tips.head(4))
```

Similar to DataFrame, you can use statistical functions on DataFrame groups. Remember that the statistical operation will be peformed on groups NOT on columns/rows.

```python
# Find average of each numeric variable, for each group (time)
print(tips_time[['total_bill', 'tip']].mean())

# Print 100 dashes
print(100*'-')

# Find max of each numeric variable, for each group (day)
print(tips_day[['total_bill', 'tip']].max())
```

```python
# To calculate multiple functions at the same time, you can use aggregate().

# Print the mean and std for each numeric variable in the group
tips_time[['total_bill', 'tip']].aggregate(['mean', 'std'])
```

```python
# Show statistics for each numeric variable
tips_time.describe()
```

```python
# Show how often a value occurs in the 'size' column, grouping by time
print(tips_time['size'].count())

# Print 50 dashes
print(50*'-')

# Show total of 'tip' column, grouping by time
print(tips_time['tip'].sum())
```

## Stacking

Joining two or more DataFrame objects is an important task in data analysis, specifically in the case of relational datasets. One can use `concat` function to join two DataFrame horizontally (i.e., columns are combined) or vertically (i.e., rows are combined) when the two DataFrame objects have the same structure. In the code cells below, we'll first split the original DataFrame into 2 DataFrames and use `concat` to vertically stack them.

```python
# Get the first 200 rows from df_tips
dfv1 = df_tips.iloc[:200]

# Get the all rows from row 200 inclusive from df_tips
dfv2 = df_tips.iloc[200:]

# Show first 5 rows of dfv1
print(dfv1.head())
print("")

# Show last 5 rows of dfv1
print(dfv1.tail(5))
print("")

# Show dimensions of dfv1
print(dfv1.shape)
print("")

# Show first 5 rows of dfv2
print(dfv2.head())
print("")

# Show last 5 rows of dfv2
print(dfv2.tail(5))
print("")

# Show dimensions of dfv2
print(dfv2.shape)
print("")
```

```python
# Vertically concatenate the above two parts
df_vertical = pd.concat([dfv1, dfv2], axis = 0)

# Show first 5 rows
print(df_vertical.head())
print("")

# Show last 5 rows
print(df_vertical.tail())
print("")

# Show dimensions
print(df_vertical.shape)
print("")

# Show statistics for all numeric variables
print(df_vertical.describe())
```

Similar to the example above, you can perform `concat` horizontally as well.

```python
# Get the first 2 columns from df_tips
dfh1 = df_tips.iloc[:, :2]

# Get the all columns from column 2 inclusive from df_tips
dfh2 = df_tips.iloc[:, 2:]

# Show first 5 rows of dfh1
print(dfh1.head())
print("")

# Show last 5 rows of dfh1
print(dfh1.tail(5))
print("")

# Show dimensions of dfh1
print(dfh1.shape)
print("")

# Show first 5 rows of dfh2
print(dfh2.head())
print("")

# Show last 5 rows of dfh2
print(dfh2.tail(5))
print("")

# Show dimensions of dfh2
print(dfh2.shape)
print("")
```

```python
# Horizontally concatenate the above two parts
df_horizontal = pd.concat([dfh1, dfh2], axis = 1)

# Show first 5 rows
print(df_horizontal.head())
print("")

# Show last 5 rows
print(df_horizontal.tail())
print("")

# Show dimensions
print(df_horizontal.shape)
print("")

# Show statistics for all numeric variables
print(df_horizontal.describe())
```

### Pivot Tables

Pivot table can be used for summarizing large dataset by grouping multiple columns. A pivot table can be created using `pivot_table` function. It takes `columns` (the columns that are displayed in the pivot table), `index` (specify row indices) and `aggfunc` (the function should be computed and displayed).


In the following code cells, we first create a pivot table on the tips dataset by first grouping the data by sex, and computing the `mean` total_bill for each day in the DataFrame.

```python
# Create a pivot table.
# Each row (index) in the pivot table will be a sex, so we are grouping by sex.
# Each column in the pivot table will be a day.
# Each cell will display the mean total_bill for a particular group (sex) and day
df_tips.pivot_table('total_bill', index = 'sex', columns = 'day')
```

```python
# Similar to above, except instead of showing mean (which is the default),
# we show median
df_tips.pivot_table('total_bill', index = 'sex', columns = 'day', aggfunc = "median")
```

The next pivot table flips this by first grouping on day and computing the mean total_bill for each sex in the DataFrame.

```python
# Similar to above, except now:
# Each row (index) in the pivot table will be a day, so we are grouping by day.
# Each column in the pivot table will be a sex.
# Each cell will display the mean total_bill for a particular group (day) and sex
df_tips.pivot_table('total_bill', index = 'day', columns = 'sex')
```

Next, the same pivot table is created but is now displayed with the aggregate function computed for all data in the appropriate row or column (via the margins argument). The next two pivot tables create hierarchical horizontal (or vertical) indices by specifying a list for the index (or columns) attribute, respectively. Finally, a pivot table with multiple summary statistics is displayed by specifying the min, max, and median functions via the aggfunc attribute during the creation of the pivot table.


```python
# Similar to above, except now:
# We include margin data for each row and column.
# i.e. We consider all days and all sexes as another group.
df_tips.pivot_table('total_bill', index = 'day', columns = 'sex', margins = all)
```

```python
# Create a pivot table.
# Each row (index) in the pivot table will be a day and smoker, so we are grouping first by day, then by smoker.
# Each column in the pivot table will be a sex.
# Each cell will display the mean total_bill for a particular group (day and smoker) and sex
df_tips.pivot_table('total_bill', index = ['day', 'smoker'], columns = 'sex')
```

```python
# Similar to above, except now:
# Each row (index) in the pivot table will be a day, so we are grouping by day.
# Each column in the pivot table will be a sex and smoker.
# Each cell will display the mean total_bill for a particular group (day) and sex and smoker
df_tips.pivot_table('total_bill', index = 'day', columns = ['sex', 'smoker'])
```

```python
# Create a pivot table.
# Each row (index) in the pivot table will be a day, so we are grouping by day.
# Each column in the pivot table will be a sex.
# We have different sets of columns to show maximum, median and minimum values.
# We include margin data for each row and column.
# "dropna = True" means that any rows with missing data will be removed.
import numpy as np
df_tips.pivot_table('total_bill', index = 'day', columns = 'sex', aggfunc = ["max", "median", "min"], margins = True, dropna = True)
```
